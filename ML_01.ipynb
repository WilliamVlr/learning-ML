```python
#Lamda function
#Menambahkan kolom baru ke dataframe dengan isi tertentu
df["kolom baru"] = 1
```

```python
df.head()
```

```python
df["kolom baru"] = df['sepal.length'] + df['sepal.width']
df.head()
```

```python
#Lamda function untuk mencari rata-rata:
df['kategori sepal length'] = df['sepal.length'].apply(lambda x: 'panjang' if x > 5.8 else 'pendek')
df.head()
```

```python
#Filtering in Pandas
df[df['kategori sepal length'] == 'panjang']
```

```python
#filter sepal length > 5 and sepal width > 3
#df[(df['sepal.length'] > 5) & (df['sepal.width'] > 3)]
df[(df['sepal.length'] > 5) & (df['sepal.width'] > 3)]
```
